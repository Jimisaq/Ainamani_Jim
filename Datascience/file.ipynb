```python
#Writing to a file
with open("isaac.txt","w") as file:
    file.write("This is the first line in my file\n")
    file.write("This is the second line in my file")
```

```python
#Appending to a file
with open("isaac.txt","a") as file:
    file.write("This is the appended line.\n")

with open("isaac.txt","r") as file:
    print(file.read())
```

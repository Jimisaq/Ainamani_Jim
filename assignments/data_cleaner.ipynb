```python
import pandas as pd

df = pd.read_csv('../Mine.csv')
df

#df.describe()
```

```python
#Removing nulls
nulless = df.dropna()
nulless
```

```python
#removing duplicate rows
nulless=nulless.drop_duplicates(keep=False)
nulless
```

```python
#Checking data types
nulless.dtypes
```

```python
#Correcting the date datatype
nulless['Date'] = pd.to_datetime(nulless['Date'],errors='coerce')
nulless=nulless.dropna(subset=['Date'])
nulless.dtypes
```

```python
nulless
```

```python
#Dropping the duration column
nulless=nulless.drop(columns=['Duration'])
nulless
```
